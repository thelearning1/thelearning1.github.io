```python
#import dependencies
import pandas as pd
import json
import pprint
import requests
```

```python
#Make an API call to enerflo.

api_endpoint_url = 'https://enerflo.io/api/v3/installs?per_page=50&page=1'


#key
key = '185461e9ebbbe28539.19676587'
headers = {'api-key':key,
          'content-type':'application/json'
          }

response = requests.get(api_endpoint_url,headers=headers)
```

```python
data = response.json()

#establish the variable to index into a given entry that the loop can iterate through
entry = data['results'][0]

pprint.pprint(entry)
```

```python
#crete the structure for the dataframe
#create the empty arrays
firstnamearr = []
lastnamearr = []
addressarr = []
statearr = []
cityarr = []
statusarr = []
agreementarr = []
installarr = []
ptoarr = []
costarr = []
sizearr = []
ppwarr = []
reparray = []

#initialize the dict to create dataframe from
report = {
    'First Name':firstnamearr,
    'Last Name':lastnamearr,
    'Address':addressarr,
    'City':cityarr,
    'State':statearr,
    'Status':statusarr,
    'Cost':costarr,
    'Size':sizearr,
    'PPW':ppwarr,
    'Sales Rep':reparray,
    'Agreement Date':agreementarr,
    'Install Date':installarr,
    'PTO Date':ptoarr
    }

#set variables to perpetuate loop and increment the page in the call
inf = 0
count = 0

#set 'while' loop to make the calls until it's done
while inf == 0:
            
    #set the value that determines the page called for
    api_endpoint_url = 'https://enerflo.io/api/v3/installs?per_page=50&page=' + str(count)

    #key
    key = '185461e9ebbbe28539.19676587'
    headers = {'api-key':key,
              'content-type':'application/json'
              }
    
    #increment the page count
    print(count)
    count = count + 1
        
    #PLACEHOLDER INCREMENTATIONs TO BREAK WHILE LOOP
    #inf = inf + 1
#     if count > 10:
#         print('limit reached, stopping query')
#         break
    
    #pull the result group into
    response = requests.get(api_endpoint_url,headers=headers)
    data = response.json()
    results = data['results']
    print(len(results))
    
    
    for entry in results:

        #pull the relevant dates
        agreement = entry['milestones'][0]['install_milestone']['completed_on']
        install_date = entry['milestones'][10]['install_milestone']['completed_on']
        pto_date = entry['milestones'][12]['install_milestone']['completed_on']

        #pull system info
        cost = entry['cost']['system_cost_net']
        size = entry['system_size']
        ppw = entry['survey']['ppw_net']

        #pull sales rep data
        sales_rep = entry['contacts'][0]['user']['name']

        #pull Customer Data
        customer = entry['customer']
        name_first = customer['first_name']
        name_last = customer['last_name']
        address = customer['address']
        city = customer['city']
        state = customer['state']
        status = customer['status']

        #append the data into the corresponding arrays
        firstnamearr.append(name_first)
        lastnamearr.append(name_last)
        addressarr.append(address)
        cityarr.append(city)
        statearr.append(state)
        statusarr.append(status)
        costarr.append(cost)
        sizearr.append(size)
        ppwarr.append(ppw)
        reparray.append(sales_rep)   
        agreementarr.append(agreement)
        installarr.append(install_date)
        ptoarr.append(pto_date)  
    
    if len(results) < 50:
        print('final page reached stopping query')
        break    
    
#generate the dataframe off of the report dict
report_df = pd.DataFrame.from_dict(report)
```
